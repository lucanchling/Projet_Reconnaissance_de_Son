```python
import scipy.io.wavfile as wavfile
import scipy.fftpack as fftpack
import numpy as np
import matplotlib.pyplot as plt
from pydub import AudioSegment
#from functions import fft
```

```python
choix_music="2"
```

```python
#s_rate, signal = wavfile.read('./music/music1.wav')
sound = AudioSegment.from_wav('./music/music'+str(choix_music)+'.wav')
sound = sound.set_channels(1)
sound.export('./music/music'+str(choix_music)+'_mono.wav', format='wav')


s_rate, signal = wavfile.read('./music/music'+str(choix_music)+'_mono.wav')
#signal = signal.astype('uint16')

#s = signal[:,0]
#création instants d'échantillons 
t = np.linspace(0, signal.shape[0]/s_rate, signal.shape[0])

#tracé du signal 
fig = plt.figure(figsize=(15,15))
plt.subplot(211)
plt.plot(signal)
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.subplot(212)
a, b, c, d = plt.specgram(signal,NFFT=256,Fs=s_rate,noverlap=128)
plt.xlabel('Time')
plt.ylabel('Frequency')
fig.colorbar(d).set_label('Intensité(dB)')

```

```python
FFT = fftpack.fft(signal)
freq = fftpack.fftfreq(signal.size,d=1/s_rate)

FFT_bis = fft(signal)
#calcul du nombre d'échantillon 
N = signal.size
FFT_abs = np.abs(FFT[:N//2])*2.0/N
freq_pos = freq[:N//2]

plt.figure(figsize=(8,8))
plt.subplot(211)
plt.plot(freq_pos, FFT_abs)

plt.subplot(212)
plt.plot(freq, FFT_bis)



```
